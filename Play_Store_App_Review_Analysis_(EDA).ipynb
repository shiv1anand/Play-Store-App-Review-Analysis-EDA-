## 3. ***Data Wrangling***

### Data Wrangling Code

Handle Duplicates in df

````python
# Write your data wrangling code to make your dataset analysis ready.
df.drop_duplicates(inplace=True)
df.duplicated().sum()
````

Handle Outlier Row in df

````python
#visualizing and identifying outliers
plt.figure(figsize=(10,4))
sns.scatterplot(data=df['Rating'])
plt.show()
df['Rating'].max()
````

````python
df = df[df['Rating'] != 19.0]
df['Rating'].max()
````

Clean and Convert 'Reviews' in df

````python
#Convert the 'Reviews' column in df to a numeric data type.
print(df['Reviews'].unique())
df['Reviews'] = pd.to_numeric(df['Reviews'])
df['Reviews'].dtype
````

Clean and Convert 'Size' in df

````python
#Clean and convert the 'Size' column in df to a numeric data type, unifying units (e.g., convert 'M' to MB and 'k' to KB, then convert everything to MB).
print(df['Size'].unique())
````

````python
# Corrected approach for 'Size' conversion:
def convert_size_to_mb(size):
    # Handle missing values
    if pd.isna(size):
        return np.nan

    size = str(size)

    # If size is in Megabytes
    if size.endswith('M'):
        return float(size.replace('M', ''))

    # If size is in Kilobytes
    elif size.endswith('k'):
        return float(size.replace('k', '')) / 1024

    # If size varies by device or unknown
    else:
        return np.nan
df['Size'] = df['Size'].apply(convert_size_to_mb)
df['Size'].dtype
````

Clean and Convert 'Installs' in df

````python
#Clean and convert the 'Installs' column in df to a numeric data type by removing '+' and ',' characters.
print(df['Installs'].unique())
````

````python
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)
df['Installs'] = pd.to_numeric(df['Installs'])
df['Installs'].dtype
````

Clean and Convert 'Price' in df

````python
#Clean and convert the 'Price' column in df to a numeric data type by removing '$' and converting to float.
print(df['Price'].unique())
````

````python
df['Price'] = df['Price'].apply(lambda x: x.replace('$', '') if '$' in str(x) else x)
df['Price'] = pd.to_numeric(df['Price'])
df['Price'].dtype
````

Handle Missing 'Rating' Values in df

````python
#Impute missing values in the 'Rating' column of df using the median.
df['Rating'] = df['Rating'].fillna(df['Rating'].median())
df['Rating'].isnull().sum()
````

Handle Missing 'Type' and 'Content Rating' in df

````python
#Impute missing values in the 'Type' and 'Content Rating' columns of df using the mode.
print(df['Type'].unique())
print(df['Type'].value_counts())
print(df['Content Rating'].unique())
print(df['Content Rating'].value_counts())
````

````python
df['Type'].fillna(df['Type'].mode()[0], inplace=True)
df['Content Rating'].fillna(df['Content Rating'].mode()[0], inplace=True)
print(df['Type'].isnull().sum())
print(df['Content Rating'].isnull().sum())
````

Convert 'Last Updated' to Datetime in df

````python
#Convert the 'Last Updated' column in df to a datetime object.
print(df['Last Updated'].dtype)
print(df['Last Updated'].head())
````

````python
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
print(df['Last Updated'].head())
df['Last Updated'].dtype
````

Handle Missing 'Current Ver' and 'Android Ver' in df

````python
#Impute missing values in 'Current Ver' and 'Android Ver' in df using the mode
print(df['Current Ver'].unique())
print(df['Current Ver'].value_counts())
print(df['Android Ver'].unique())
print(df['Android Ver'].value_counts())
````

````python
df['Current Ver'].fillna(df['Current Ver'].mode()[0], inplace=True)
df['Android Ver'].fillna(df['Android Ver'].mode()[0], inplace=True)
print(df['Current Ver'].isnull().sum())
print(df['Android Ver'].isnull().sum())
````

Handle Duplicates and Missing Values in df2

````python
#Remove duplicate rows from df2. Then, remove rows from df2 where 'Translated_Review' is null.
df2.drop_duplicates(inplace=True)
df2.duplicated().sum()
````

````python
df2.dropna(subset=['Translated_Review'], inplace=True)
df2['Translated_Review'].isnull().sum()
````
